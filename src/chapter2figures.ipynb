```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn

%matplotlib inline
```

```python
esco = np.linspace(0,17,100)
minimo = 500
salario1 = minimo + 100 * esco 
salario2 = minimo + 100 * esco**2 
salario3 = minimo - 300 + 100 * esco 
salario4 = minimo - 300 + 30 * esco 
```

```python
plt.plot(esco,salario1,'black')
plt.xlabel('Escolaridad')
plt.ylabel('Salario')
plt.savefig("../img/capitulo2/modelo1.png")
```

```python
plt.plot(esco,salario2,'black')
plt.xlabel('Escolaridad')
plt.ylabel('Salario')
plt.savefig("../img/capitulo2/modelo2.png")
```

```python
plt.plot(esco,salario3,'pink',label = 'Mujer')
plt.plot(esco,salario1,'lightblue', label = 'Varon')
plt.xlabel('Escolaridad')
plt.ylabel('Salario')
plt.legend(loc=0)
plt.savefig("../img/capitulo2/modelo3.png")
```

```python
plt.plot(esco,salario4,'pink',label = 'Mujer')
plt.plot(esco,salario1,'lightblue', label = 'Varon')
plt.xlabel('Escolaridad')
plt.ylabel('Salario')
plt.legend(loc=0)
plt.savefig("../img/capitulo2/modelo4.png")
```
