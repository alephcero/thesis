```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn

%matplotlib inline
```

```python
esco = np.linspace(0,17,100)
minimo = 500
salario1 = minimo + 100 * esco 
salario2 = minimo + 100 * esco**2 
salario3 = minimo - 300 + 100 * esco 
salario4 = minimo - 300 + 30 * esco 
```

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.plot(esco,salario1,'black')
ax1.set_xlabel('Escolaridad')
ax1.set_ylabel('Salario')

ax2.plot(esco,salario2,'black')
ax2.set_xlabel('Escolaridad')
ax2.set_ylabel('Salario')

ax3.plot(esco,salario3,'pink',label = 'Mujer')
ax3.plot(esco,salario1,'lightblue', label = 'Varon')
ax3.set_xlabel('Escolaridad')
ax3.set_ylabel('Salario')
ax3.legend(loc=0)



ax4.plot(esco,salario4,'pink',label = 'Mujer')
ax4.plot(esco,salario1,'lightblue', label = 'Varon')
ax4.set_xlabel('Escolaridad')
ax4.set_ylabel('Salario')
ax4.legend(loc=0)

```
